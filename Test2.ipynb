```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Update.csv')
df.set_index('Country', inplace=True)
df.drop(['Country'], inplace=True)
male_df = df.copy()[df['Sex'].str.contains('Male', case=True)]
plt.figure()

male_df['Total'] = male_df['Year 2016'] + male_df['Year 2015'] + male_df['Year 2010'] + male_df['Year 2010']
male_df.sort_values(by=['Total'], inplace=True)
exclude = ['Total']

fig, ax = plt.subplots(ncols=2, figsize=(15,4))
lowMale_df = male_df[male_df.columns.difference(exclude)].head(15)
plotLow = lowMale_df.plot(kind='bar',title='Top 15 Countries with lowest total male suicide',colormap='spring',ax=ax[0])
plotLow.set_ylabel('Suicides per 100k')
plotLow.set_ylim(0, 80)

highMale_df = male_df[male_df.columns.difference(exclude)].tail(15)
plotHigh = highMale_df.plot(kind='bar',colormap='viridis',title='Top 15 Countries highest total male suicide',ax=ax[1])
plotHigh.set_ylabel('Suicides per 100k')
plotHigh.set_ylim(0, 87)
```
